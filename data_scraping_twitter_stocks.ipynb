```python
from ntscraper import Nitter
```

```python
scraper=Nitter(0)
```

```python
tweets = scraper.get_tweets('$MSFT', mode='hashtag', number=15, language='en', since='2020-01-01')

if tweets:
    # Process the tweets
    print("Tweets fetched successfully")
else:
    print("No tweets found for the given query")
```

```python
final_tweets=[]
for x in tweets['tweets']:
  data=[x['date'],x['user']['profile_id'],x['text'],x['user']['username']]
  final_tweets.append(data)
```

```python
scrape=pd.DataFrame(final_tweets,columns=['Datetime','ID','text','username'])
```
